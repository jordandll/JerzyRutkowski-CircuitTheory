# Hyperbolic Functions
In mathematics, **hyperbolic functions** are analogues of the ordinary trigonometric functions, but defined using the hyperbola rather than the circle. Just as the points $(\cos t, \sin t)$ form a circle with a unit radius, the points $(\cosh t, \sinh t)$ form the right half of the unit hyperbola.  The unit hyperbola is given by:  

\begin{equation}\tag{1}
    x^2-y^2=1
\end{equation}

```python
x = np.linspace(-3,3,num=200)
x_tanh = np.linspace(-8, 8, num=200)

fig, ax = plt.subplots()
ax.set_title('Hyperbolic Functions', size=13)
ax.set_xlim(-8,8)

ax.plot(x, np.sinh(x), label=r'$\sinh x$')
ax.plot(x, np.cosh(x), label=r'$\cosh x$')
ax.plot(x_tanh, np.tanh(x_tanh), label=r'$\tanh x$')
ax.legend()
```

Via the above equation and equation (6b) in the [conics](Conics-Rectilinear.ipynb#Hyperbola) chapter we can find the parameters, in term of eccentricity, denoted as '$\mathrm{e}$', focus and directrix, of the unit hyperbola.  

\begin{equation}\tag{1}
    \text{a: }\mathrm{e} = \sqrt{2}\quad\quad\text{and}\quad\quad\text{b: }p = \frac{1}{\sqrt{2}}
\end{equation}

Note that, in the referenced chapter/notebook, the eccentricity is denoted as '$e$', but that denotation conflicts with the standard denotation of Euler's constant used herein.

```python
# Create the x-domain.
x = np.linspace(1, 3, num=100)
x_f = np.linspace(-3,3,num=100)

# Define the function that gives the top half of the unit hyperbola and the assymptote.
y_p = np.sqrt(x**2-1)
f = x_f

# Create the figure and axes objects.
fig, ax = plt.subplots()
ax.set_title('Unit Hyperbola', size=13)

# Plot the unit hyperbola.
ax.plot(x, y_p, x, -y_p, -x, y_p, -x, -y_p, color='blue')
ax.plot(y_p, x, -y_p, x, y_p, -x, -y_p, -x, color='red')
ax.plot(x_f, f, x_f, -f)
```

The **assymptotes** of the unit hyperbola are given by:  

$$ y = \pm x$$

Note that the unit hyperbola is parametrically defined as:  

\begin{equation}\tag{3}
    (\sec\theta, \tan\theta) \quad\quad\text{and}\quad\quad(\pm\cosh a, \sinh a)
\end{equation}

```python
def csch(x):
    return 1/np.sinh(x)
def sech(x):
    return 1/np.cosh(x)
def coth(x):
    return 1/np.tanh(x)
```

```python
x = np.linspace(-5,5,num=200)
fig, ax = plt.subplots()
ax.set_title('Hyperbolic Functions', size=13)
ax.set_ylim(-8, 8)

ax.plot(x, csch(x), label=r'csch x')
ax.plot(x, sech(x), label=r'sech x')
ax.plot(x, coth(x), label=r'coth x')
ax.legend()
```

**Proposition**:  Rotating the unit hyperbola by $45^\circ$ gives $\text{csch}\, x$.

```python
theta = np.linspace(-np.radians(44), np.radians(44), num=200)
theta += np.radians(45)
r = 1/np.sqrt(1-2*np.sin(theta-np.radians(45))**2)
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.set_title('Unit Hyperbola', size=13)

ax.plot(theta, r, theta + np.pi, r)
```

**Proof**:  Let $r^*(\theta)$ be defined as $r^*(\theta):=r(\theta-45^\circ)$.  It now follows that:  

$$ r^* = \sqrt{x^2+y^2} = \frac{1}{\sqrt{1-2\sin^2\big(\theta - 45^\circ\big)}}$$

$$ x^2 + y^2 = \frac{1}{1-2\sin^2\big(\theta - 45^\circ\big)}$$

$$\sin^2\big(\theta-45^\circ\big) = \left(\frac{1}{\sqrt{2}}\,\sin\theta - \frac{1}{\sqrt{2}}\cos\theta\right)^2 = 0.5\big(\sin\theta - \cos\theta\big)^2 = 0.5\big(1- 2\sin\theta\cos\theta\big) $$

$$ x^2 + y^2 = \frac{1}{2\sin\theta\cos\theta} = \frac{1}{2}\frac{r^{*2}}{xy} = 0.5\,\frac{x^2+y^2}{xy}$$

Thus,  

$$ y = \frac{1}{2x}$$

```python
x = np.linspace(-5,5,num=200)
y = 1/(2*x)
fig, ax = plt.subplots()
ax.set_title('Rotation', size=14)
ax.set_ylim(-8,8)

ax.plot(x, y, label=r'$0.5x^{-1}$')
ax.plot(x, csch(x), label=r'csch $x$', ls='dashed')
ax.plot(x, coth(x), label=r'coth $x$', ls='dotted')
ax.legend()
```

## Hyperbolic Angle
The hyperbolic functions take a real argument called a hyperbolic angle. The size of a hyperbolic angle is twice the area of its hyperbolic sector. The hyperbolic functions may be defined in terms of the legs of a right triangle covering this sector.  

![hyperbola-sector](https://upload.wikimedia.org/wikipedia/commons/thumb/b/bc/Hyperbolic_functions-2.svg/310px-Hyperbolic_functions-2.svg.png)

A ray through the unit hyperbola $x^2  y^2 = 1$ at the point $(\cosh a, \sinh a)$, where $a$ is twice the area between the ray, the hyperbola, and the $x$-axis. For points on the hyperbola below the $x$-axis, the area is considered negative (see animated version with comparison with the trigonometric (circular) functions)

Now let $f(x):=mx$ and $g(x):= \sqrt{x^2-1}$.  

$$ 0.5a = 0.5a(x) = \int_0^x f(x)\,\mathrm{d}x - \int_0^x g(x)\,\mathrm{d}x$$

$$ \int_0^xf(x)\,\mathrm{d}x = 0.5mx^2$$

Since $g(x)$ is hard to integrate, let's work with the rotated unit hyperbola to find the area.  Let $h(x)$ be defined as:  

$$ h(x):= y = \frac{1}{2x}$$

Now, suppose there exists an $x^*,y^*$-axes that is given by rotating the $x,y$-axes by $45^\circ$.  It now follows that:  

$$ x = x^*\cos45^\circ = \frac{x^*}{\sqrt{2}}$$

$$ m:=m(x^*) = \frac{g(x^*)}{x^*} = \frac{\sqrt{x^{*2}-1}}{x^*}$$

$$ f(x) = \tan(\arctan m(x^*) + 45^\circ) x$$

## Hyperbolic Sector

```python
x_ast = 2
m = np.sqrt(x_ast**2-1)/x_ast

A_x, A_y = 1/np.sqrt(2), 1/np.sqrt(2)
B_x = (2*np.sqrt(2)-np.sqrt(6))/2
B_y = 2*np.sqrt(2) - B_x
B_x2 = (2*np.sqrt(2)+np.sqrt(6))/2
B_y2 = 2*np.sqrt(2) - B_x2

x = np.linspace(0.0001,5,num=200)
f = np.tan(np.arctan(m) + 0.25*np.pi)*x
f_ast = np.tan(-np.arctan(m) + 0.25*np.pi)*x
h = 1/(2*x)
alt = np.sqrt(2)*x_ast - x

fig, ax = plt.subplots()
ax.set_title('Hyperbolic Sector', size=13)
ax.set_ylim(0,5)

ax.plot(x, x, x, alt)
ax.plot(x, f, x, f_ast)
ax.plot(x, h)
ax.plot([A_x for i in range(25)], np.linspace(0, 5, num=25), ls='dashed')
ax.plot([B_x for i in range(25)], np.linspace(0, 5, num=25), ls='dashed')
ax.annotate('A', (A_x, A_y))
ax.annotate('B', (B_x, B_y))
ax.annotate('B\'', (B_x2, B_y2))
```

$$ A = \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right)$$  

And, $B$ is the intersection between the green, purple (hyperbola), and orange lines, which are given by:  

$$ y = f(x) \quad\quad\text{and}\quad\quad y = h(x) = \frac{1}{2x} \quad\quad\text{and}\quad\quad y = 2\sqrt{2} - x$$

respectively.  

$$ x = 2\sqrt{2} - \frac{1}{2x} \quad\implies\quad 2x^2 - 4\sqrt{2}x + 1 = 0 $$

$$ x = \frac{4\sqrt{2}\pm\sqrt{32-8}}{4} = \frac{2\sqrt{2}\pm \sqrt{6}}{2}$$

Thus,  

$$ B_x = \frac{2-\sqrt{3}}{\sqrt{2}} \quad\quad\text{and}\quad\quad B_y = 2\sqrt{2} - B_x$$

$$ B'_x = \frac{2\sqrt{2}+\sqrt{6}}{2} \quad\quad\text{and}\quad\quad B'_y = 2\sqrt{2}-B'_x$$

Parametrically,  

$$ y = x^*\sqrt{2} - x$$

$$ x = x^*\sqrt{2} - \frac{1}{2x}\quad\implies\quad 2x^2 - 2x^*\sqrt{2}x + 1 = 0$$

$$ x = \frac{2x^*\sqrt{2}\pm\sqrt{8x^{*2} - 8}}{4} = \frac{x^*\pm\sqrt{x^{*2}-1}}{\sqrt{2}}$$

$$ B_x(x^*) = \frac{x^*-\sqrt{x^{*2}-1}}{\sqrt{2}} \quad\quad\text{and}\quad\quad B_y(x^*) = x^*\sqrt{2} - B_x(x^*)$$

$$ B'_x(x^*) = \frac{x^*+\sqrt{x^{*2}-1}}{\sqrt{2}} \quad\quad\text{and}\quad\quad B'_y(x^*) = x^*\sqrt{2} - B'_x(x^*)$$

The **hyperbolic angle** is parametrically given by:  

\begin{equation}
    0.5a:= 0.5a(x^*) = \int_0^{\beta}f(x)\,\mathrm{d}x - \int_0^{\beta}x\,\mathrm{d}x + \int_{\beta}^{\alpha}h(x)\,\mathrm{d} - \int_{\beta}^{\alpha}x\,\mathrm{d}x
\end{equation}

Multiplying throughout the above equation by $2$ gives:  

$$a = a(x^*) = B_x(x^*)B_y(x^*) - 0.5 - 0.5\log2-\log B_x(x^*)$$

$$ = B_x(x^*)x^*\sqrt{2} - B_x^2(x^*) - 0.5 - \log\big(x^*-\sqrt{x^{*2}-1}\big)$$

$$ = x^*\big(x^*+\sqrt{x^{*2}-1}\big) - 0.5\big(x^*-\sqrt{x^{*2}-1}\big)^2 - 0.5 - \log\big(x^*-\sqrt{x^{*2}-1}\big)$$

$$ = x^*\big(x^*+\sqrt{x^{*2}-1}\big) - 0.5\big(x^{*2} + x^{*2} - 1 - 2x^*\sqrt{x^{*2}-1}\big) - 0.5 - \log\big(x^*-\sqrt{x^{*2}-1}\big)$$

Finally, the hyperbolic angle is given by:  

\begin{equation}\tag{3}
    a = a(x^*) = - \log\big|x^*-\sqrt{x^{*2}-1}\big|\quad\quad\text{where }x^*\ge 1
\end{equation}

Now let $p(x^*)$ be defined as the argument to the natural logarithm function in the above equation, with the caveat of said equation.  

$$ p'(x^*) = 1 - \frac{x^*}{\sqrt{x^{*2}-1}} \quad\quad\text{where }x^*\ge 1$$

The above function has no roots and thus $p(x^*)$ never changes direction.  Now, which direction are we dealing with?  

$$ 1 \overset{?}{>} \frac{x^*}{\sqrt{x^{*2}-1}}$$

$$ -1\overset{?}{>}0$$

The above implies that $p(x^*)$ goes in the negative direction.  And, since the domain of $p(x^*)$ starts at $1$ and $p(1) = 1$, which corresponds to $a(1)=0$,  

$$ a(x^*) > 0 \quad\quad\text{for }x^*>1$$

and 

$$ 1>p(x^*) > 0 \quad\quad\text{for }x^* > 1$$

# Definitions

Herein are some relationships that may define the hyperbolic functions

## Exponential Definitions
Recall that:  

\begin{equation}\tag{4}
    \text{a: }\;\cosh a = x^* \quad\quad\text{and}\quad\quad \text{b: }\;\sinh a = \pm g(x^*) = \pm \sqrt{x^{*2}-1}
\end{equation}

To find the hyperbolic functions in terms of $a$, we must first solve for $x^*$ in terms of $a$ via equation (3).  

$$ x^* - \sqrt{x^{*2}-1} = e^{-a}$$

$$ x^{*2}-1 = e^{-2a} + x^{*2} - 2e^{-a}x^*$$

$$ 2e^{-a}x^* = e^{-2a} + 1$$

\begin{equation}\tag{5}
    x^* = \frac{e^{-a} + e^a}{2}
\end{equation}

### Hyperbolic Cosine
Via equations (5) and (4a) and substitution,  

\begin{equation}\tag{4a}
    \cosh a = \frac{e^{-a} + e^a}{2}
\end{equation}

And, via equations (4b) and (5) and substitution,  

$$ g(a) = \sqrt{\frac{e^{-2a} + 2 + e^{2a}}{4} - 1}$$

$$ = 0.5\sqrt{e^{-2a} - 2 + e^{2a}}$$

$$ = \frac{\big|e^a - e^{-a}\big|}{2}$$

Thus,  

$$ \pm g(a) = \frac{e^a - e^{-a}}{2}$$

### Hyperbolic Sine
And finally,  

\begin{equation}\tag{4b}
    \sinh a = \frac{e^a - e^{-a}}{2}
\end{equation}

## Hyperbola  

As has already been stated, hyperbolic sine and cosine can be used to parametrically define a hyperbola.  To verify:  Let $a:=\psi$.  Via the parametric definition of the unit hyperbola,  

\begin{equation}\tag{6a}
    \cosh^2a - \sinh^2a = 1
\end{equation}

$$ e^{-2a} + 2 + e^{2a} - e^{2a} - e^{-2a} + 2 = 4$$

$$ 4 = 4$$

Q.E.D.

Equation (6a) can be further generalized to all hyperbolas given by:  

$$ \frac{\cosh^2a}{A^2} - \frac{\sinh^2a}{B^2} = 1$$

$$ B^2\cosh^2a - A^2\sinh^2a = (AB)^2$$

$$ B^2\big(e^{2a} + e^{-2a} + 2\big) - A^2\big(e^{2a} + e^{-2a} - 2\big) = (2AB)^2$$

The **hyperbolic angle** is thus given by:

# Properties

## Hyperbolic Cosine

It can be shown that the area under the curve of the hyperbolic cosine (over a finite interval) is always equal to the arc length corresponding to that interval.  

Let $y:= f(x)=\cosh x$.  It then follows that:  

$$ \mathrm{d} y = \sinh x\,\mathrm{d}x$$

The aforementioned arc length from $b$ to $a$ is given by:  

$$ \int_a^b \sqrt{\mathrm{d}y^2 + \mathrm{d}x^2}$$

And, the area under the curve given by $f(x)$ in that same interval is given by:  

$$ \int_a^b y\,\mathrm{d}x$$

The **proposition** is thus:  

\begin{equation}\tag{8}
    \int_a^b y\,\mathrm{d}x = \int_a^b\sqrt{\mathrm{d}y^2 + \mathrm{d}x^2}
\end{equation}

**Proof**:  

$$ \int_a^b\sqrt{\mathrm{d}y^2 + \mathrm{d}x^2} = \int_a^b\sqrt{\sinh^2x + 1}\,\mathrm{d}x = \int_a^b \cosh x\,\mathrm{d}x$$

And,  

$$ \int_a^b y\,\mathrm{d}x = \int_a^b \cosh x\,\mathrm{d}x$$

Via the above two equations and equivalence principle we have equation (8).  Q.E.D.
